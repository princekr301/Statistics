# CDF
### Cumulative Distribution Function (CDF)
The Cumulative Distribution Function (CDF) is a way to show the probability that a random variable 
 will be less than or equal to a certain value . 

It essentially adds up all the probabilities up to that point.

Think of it as a running total of probabilities. For example, if you have a six-sided die, the CDF at 3 would be the total probability of rolling a 1, 2, or 3, which is 50% (since there's a 1/6 chance for each outcome).

The CDF helps you understand the overall distribution of the values a random variable can take. It starts at 0 and increases to 1 as you move along the possible values.

```python
import numpy as np
import pandas as pd
```

```python
data = []
for i in range(1000):
    dice = np.random.randint(1,7)
    data.append(dice)
```

```python
df = (pd.Series(data).value_counts()/pd.Series(data).value_counts().sum()).sort_index()
df
```

```python
# Cumulative sum
x  = np.cumsum(df)
```

```python
x.plot(kind = "bar");
```

# Insights :
1. Its a example of Cumulative distribution Function
2. Non-decreasing: The CDF is a non-decreasing function.
3. Range: The CDF, (), takes values in the interval [0,1]

# PDF

### Probability Density Function

PDF stands for Probability Density Function. It's used to specify the probability of a continuous random variable falling within a particular range of values, as opposed to taking on any one value. The PDF is essentially the function that describes the relative likelihood of a random variable assuming a given value.
